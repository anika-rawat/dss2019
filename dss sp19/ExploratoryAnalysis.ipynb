```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_seq_items = 2000
pd.set_option('display.max_columns', 60)
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_csv('EdStatsData.csv')
brit = data[data['Country Name'] == 'United Kingdom']
row_counts = brit.apply(lambda x: x.count(), axis=1)
brit['Row Count'] = row_counts
brit = brit[brit['Row Count'] > 30]
print(brit['Indicator Name'])
```

```python
years = [x for x in range(1985,2015)]
post_85 = brit.drop([str(x) for x in range(1970,1985)],axis=1)
post_85 = post_85.drop(post_85.columns[post_85.columns.get_loc('2015'):len(post_85.columns)-1], axis=1)
post_85
```

```python
male_enroll = post_85[post_85['Indicator Name']=='Adjusted net enrolment rate, lower secondary, female (%)'].iloc[:,4:len(post_85.columns)-1]
female_enroll = post_85[post_85['Indicator Name']=='Adjusted net enrolment rate, lower secondary, male (%)'].iloc[:,4:len(post_85.columns)-1]
plt.plot(years,female_enroll.iloc[0,:],label='Female')
plt.plot(years,male_enroll.iloc[0,:], label='Male')
plt.title('Male vs Female Net Enrollment')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(loc='lower right')
```

```python
fem_enroll = post_85[post_85['Indicator Name']=='Enrolment in tertiary education per 100,000 inhabitants, female'].iloc[:,4:len(post_85.columns)-1]
male_enroll = post_85[post_85['Indicator Name']=='Enrolment in tertiary education per 100,000 inhabitants, male'].iloc[:,4:len(post_85.columns)-1]
plt.plot(years,male_enroll.iloc[0,:],label='Male')
plt.plot(years,fem_enroll.iloc[0,:], label='Female')
plt.title('Male vs Female Tertiary Enrollment')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(loc='lower right')
```

```python
fem_noschool = post_85[post_85['Indicator Name']=='Rate of out-of-school children of primary school age, female (%)'].iloc[:,4:len(post_85.columns)-1]
male_noschool = post_85[post_85['Indicator Name']=='Rate of out-of-school children of primary school age, male (%)'].iloc[:,4:len(post_85.columns)-1]
plt.plot(years, fem_noschool.iloc[0,:], label='Female')
plt.plot(years,male_noschool.iloc[0,:], label='Male')
plt.title('Male vs Female Rate of out-of-school students')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(loc='lower right')
```

```python
male_tertiary = post_85[post_85['Indicator Name']=='School life expectancy, tertiary, male (years)'].iloc[:,4:len(post_85.columns)-1]
fem_tertiary = post_85[post_85['Indicator Name']=='School life expectancy, tertiary, female (years)'].iloc[:,4:len(post_85.columns)-1]
plt.plot(years, male_tertiary.iloc[0,:], label='Male')
plt.plot(years, fem_tertiary.iloc[0,:], label='Female')
plt.title('Tertiary Education Expectancy male vs female')
plt.xlabel('Year')
plt.ylabel('Years')
plt.legend(loc='lower right')
```

```python
enrollment = post_85[(post_85['Indicator Name']=='Enrolment in Grade 1 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 2 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 3 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 4 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 5 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 6 of primary education, both sexes (number)')|
                 (post_85['Indicator Name']=='Enrolment in Grade 7 of primary education, both sexes (number)')].iloc[:,4:len(post_85.columns)-1]                                                                                                
labels=['Grade 1','Grade 2', 'Grade 3', 'Grade 4', 'Grade 5', 'Grade 6', 'Grade 7']
y = [enrollment.iloc[0, :], enrollment.iloc[1, :], enrollment.iloc[2, :], enrollment.iloc[3, :], enrollment.iloc[4, :], enrollment.iloc[5, :],enrollment.iloc[6, :]]
plt.stackplot(years, y, labels=labels)
plt.title('Enrollment in All Years of Primary School')
plt.xlabel('Years')
plt.ylabel('Number of Students (in 1e8)')
plt.legend(loc='lower right')
```

```python
teachers = post_85[post_85['Indicator Name']=='Pupil-teacher ratio in primary education (headcount basis)'].iloc[:,4:len(post_85.columns)-1]
expectancy = post_85[post_85['Indicator Name']=='School life expectancy, primary, both sexes (years)'].iloc[:,4:len(post_85.columns)-1]
# print(len(teachers.iloc[0:1,:]), len(expectancy.iloc[0:1,:]))
plt.scatter(teachers, expectancy, color='green')
plt.title('Number of Teachers vs School Life Expectancy')
plt.xlabel('Number of Teachers')
plt.ylabel('School life expectancy')
plt.legend(loc='lower right')
```

```python
gdp = post_85[post_85['Indicator Name']=='GDP per capita (current US$)'].iloc[:,4:len(post_85.columns)-1]
secondary = post_85[post_85['Indicator Name']=='Enrolment in secondary education, both sexes (number)'].iloc[:,4:len(post_85.columns)-1]
plt.scatter(gdp, secondary, color='red')
plt.title('GDP Per Capita vs Enrollment in Secondary Education')
plt.xlabel('Enrollment in Secondary Education')
plt.ylabel('GDP Per Capita')
plt.legend(loc='lower right')
```

```python
general = post_85[post_85['Indicator Name']=='Percentage of female students in secondary education enrolled in general programmes, female (%)'].iloc[:,4:len(post_85.columns)-1]
labor = post_85[post_85['Indicator Name']=='Labor force, female (% of total labor force)'].iloc[:,4:len(post_85.columns)-1]
plt.scatter(general, labor, color='blue')
plt.title('Percent of female students in secondary education enrolled in general programs vs Percent of females in work force')
plt.xlabel('Percent in General Programs')
plt.ylabel('Percent in work force')
plt.legend(loc='lower right')
```
